```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)

student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
##Calculate the Total School Budget by District
school_data_groupbytype = school_data.groupby(['type'])
final_df = school_data_groupbytype['budget'].sum()
final_df = final_df.to_frame()
final_df.reset_index()
```

```python
##Calculate the total Number of Schools by District and add it to the final df
dataf = school_data_groupbytype['School ID'].count()
final_df = final_df.assign(School_count=dataf.values)
final_df.reset_index()
final_df
```

```python
##Calculate the math mean scores
school_data_complete_grpbydistricttype = school_data_complete.groupby(["type"])
scores_df = school_data_complete_grpbydistricttype['math_score'].mean()
final_df = final_df.assign(math_scores = scores_df.values)                                                 
scores_df = school_data_complete_grpbydistricttype['reading_score'].mean()
final_df = final_df.assign(reading_scores = scores_df.values)
final_df
```

```python
##Overall Avg Scores
final_df['OverallAvg_Scores']= (final_df['math_scores'] + final_df['reading_scores'])/2
final_df
```

```python
##Student Count 
#print(final_df)
final_df = final_df.assign(Student_Pop =school_data_complete_grpbydistricttype["size"].count()) 
final_df
```

```python
## Math Pass Percentage for Charter and District Students
MathPassCnt_C= school_data_complete[((school_data_complete.loc[:, 'math_score'] > 69) &\
                                     (school_data_complete.loc[:,'type'] == 'Charter'))]
MathPassCnt_C= MathPassCnt_C['Student ID'].count()
totalcnt = final_df.loc['Charter','Student_Pop']
CharterMathPassPerct = round(100* (MathPassCnt_C/totalcnt),2)
MathPassCnt_D= school_data_complete[((school_data_complete.loc[:, 'math_score'] > 69) &\
                                     (school_data_complete.loc[:,'type'] == 'District'))]
MathPassCnt_D= MathPassCnt_D['Student ID'].count()
totalcnt = final_df.loc['District','Student_Pop']
DistrictMathPassPerct = round(100* (MathPassCnt_D/totalcnt),2)
final_df['MathPass%'] = [CharterMathPassPerct,DistrictMathPassPerct]
final_df
```

```python
## Reading Pass Percentage for Charter and District Students
ReadingPassCnt_C= school_data_complete[((school_data_complete.loc[:, 'reading_score'] > 69) & (school_data_complete.loc[:,'type'] \
                                                                          == 'Charter'))]
ReadingPassCnt_C= ReadingPassCnt_C['Student ID'].count()

totalcnt = final_df.loc['Charter','Student_Pop']
CharterReadingPassPerct = round(100* (ReadingPassCnt_C/totalcnt),2)

ReadingPassCnt_D= school_data_complete[((school_data_complete.loc[:, 'reading_score'] > 69) & (school_data_complete.loc[:,'type'] \
                                                                          == 'District'))]
ReadingPassCnt_D= ReadingPassCnt_D['Student ID'].count()
totalcnt = final_df.loc['District','Student_Pop']
DistrictReadingPassPerct = round(100* (ReadingPassCnt_D/totalcnt),2)

######District Summary####################
final_df['ReadPass%'] = [CharterReadingPassPerct,DistrictReadingPassPerct]
final_df.round(2).style.highlight_max(subset=['MathPass%','ReadPass%'], color='yellow', axis=0)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#school_data summary
schsummary_df = school_data_complete.groupby(["type","school_name"])
df2= schsummary_df['math_score','reading_score'].mean()
schsummary_df= schsummary_df["Student ID"].count().reset_index()
schsummary_df.rename(columns={'Student ID': 'Total Students'}, inplace= True)
schsummary_df = pd.merge(schsummary_df, school_data, how="left", on=["school_name", "school_name"])
schsummary_df['Per Student Budget']= schsummary_df['budget'] /schsummary_df['Total Students'] 
schsummary_df.rename(columns={'type_x': 'type'}, inplace= True)
schsummary_df.drop(schsummary_df.columns[[3,4,5]], axis=1, inplace=True)
schsummary_df = pd.merge(schsummary_df, df2, how="left", on=["school_name", "school_name"])
schsummary_df.rename(columns={'reading_score':'Avgreading_score'}, inplace = True)
schsummary_df.rename(columns={'math_score':'Avgmath_score'}, inplace = True)
schsummary_df.round(2)
```

```python
## Calculations on Avg Pass and Fail for each School
# first Reading
ReadingPassCnt_C= school_data_complete[(school_data_complete.loc[:,'reading_score'] > 69)] 
ReadingPassCnt_C = ReadingPassCnt_C.groupby('school_name')
ReadingPassCnt_C['Student ID'].count()
ReadingPassCnt_df = ReadingPassCnt_C['Student ID'].count().to_frame()
ReadingPassCnt_df.rename(columns={'Student ID': 'ReadingPass'}, inplace= True)
schsummary_df = pd.merge(schsummary_df, ReadingPassCnt_df, how="left", on=["school_name", "school_name"])
schsummary_df["ReadingPass%"] = round(100 * (schsummary_df["ReadingPass"]/schsummary_df["Total Students"]),2)
schsummary_df
```

```python
# Then Math
MathPassCnt_C= school_data_complete[(school_data_complete.loc[:,'math_score'] > 69)] 
MathPassCnt_C = MathPassCnt_C.groupby('school_name')
MathPassCnt_C['Student ID'].count()
MathPassCnt_df = MathPassCnt_C['Student ID'].count().to_frame()
MathPassCnt_df.rename(columns={'Student ID': 'MathPass'}, inplace= True)
schsummary_df = pd.merge(schsummary_df, MathPassCnt_df, how="left", on=["school_name", "school_name"])
schsummary_df["MathPass%"] = round(100 * (schsummary_df["MathPass"]/schsummary_df["Total Students"]),2)
schsummary_df
```

```python
#Overall Passing Rate (Average of the above two)
schsummary_df["OverallPass%"] = round(((schsummary_df["ReadingPass%"] + schsummary_df["MathPass%"])/2),2)
schsummary_df.reset_index()
#Cleanup Drop Math/Reading Count Cols
schsummary_df.drop(schsummary_df.columns[[7,9]], axis=1, inplace=True)
schsummary_df.round(2)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python

schsummary_df.sort_values(by=['OverallPass%'], ascending=False, inplace=True)
schsummary_df.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
schsummary_df.sort_values(by=['OverallPass%'], ascending=True ,inplace=True)
schsummary_df.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#9th Math
grade_df = school_data_complete[(school_data_complete.loc[:,'grade'] == "9th")] 
grade_df = grade_df.groupby('school_name')
grade_df = grade_df['math_score'].mean().to_frame()
grade_df = grade_df.assign(Math9GradeAvg=grade_df.values)
grade_df.drop(grade_df.columns[[0]], axis=1, inplace=True)
#10th Math
grade10_df = school_data_complete[(school_data_complete.loc[:,'grade'] == "10th")] 
grade10_df = grade10_df.groupby('school_name')
grade10_df = grade10_df['math_score'].mean().to_frame()
grade_df = grade_df.assign(Math10GradeAvg=grade10_df.values)
#11th Math
grade11_df = school_data_complete[(school_data_complete.loc[:,'grade'] == "11th")] 
grade11_df = grade11_df.groupby('school_name')
grade11_df = grade11_df['math_score'].mean().to_frame()
grade_df = grade_df.assign(Math11GradeAvg=grade11_df.values)
#12th Math
grade12_df = school_data_complete[(school_data_complete.loc[:,'grade'] == "12th")] 
grade12_df = grade12_df.groupby('school_name')
grade12_df = grade12_df['math_score'].mean().to_frame()
grade_df = grade_df.assign(Math12GradeAvg=grade12_df.values)
display(HTML(grade_df.round(2).to_html()))

```
